```python
# FINAL CODE

# Import libraries
import pandas as pd
import numpy as np

# Import salmon catch data
catch_data = pd.read_csv("data/salmon_data.csv")

# Update the value in 'Catch' column at row positon 400 to 1
catch_data["Catch"].at[400] = 1

# 1. Update "Catch" values to be type int64
catch_data["Catch"] = catch_data["Catch"].astype(np.int64)

# Create a horizontal bar plot with mean salmon catch (1878-1997) by region
catch_data.groupby("Region")["Catch"].mean().sort_values(ascending=False).plot(kind = "barh", # Horizontal bar plot
                             title = "Average Alaskan Salmon Catch by Region (1878-1997)", # Add title
                             xlabel = "Catch (thousands of fish)", # Specify x label 
                             color = "salmon") # Color bars
```
